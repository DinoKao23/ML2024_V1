Handle Missing Value

```python
# Make a copy of the dataset for future comparison
df_fix = df.copy()
```

```python
# Inspect the missing value in the dataframe
df_fix.isna().sum().sort_values(ascending=False)
```

```python
# Missing value Percent to see the percent in dataframe
na_percent = df_fix.isna().mean().round(4) * 100
na_percent.sort_values(ascending=False)
```

```python
# Drop columns with 40% na or more 
df_fix = df_fix.drop(['PoolQC', 'MiscFeature','Alley', 'Fence', 'FireplaceQu'], axis = 1)
df_fix.shape
```

```python
# There are 14 columns with missing values 
na_col = df_fix.columns[df_fix.isna().any()]
df.loc[:, na_col].shape
```

```python
# Find the string columns still has missing values
na_col = df_fix.columns[df_fix.isna().any()]
missing_na_cat = df_fix.loc[:,na_col].select_dtypes('object')
missing_na_cat
```

```python
# Finding columns with missing datatype (string or numeric)
missing_na_num = df_fix.loc[:,na_col].select_dtypes(['int','float'])
missing_na_num
```

```python
# Use mode to fill in Categorical Missing Value, and median to fill in Numerical Missing Value
imputer_num = SimpleImputer(strategy = 'median')
imputer_cat = SimpleImputer(strategy = 'most_frequent')

# impute the num column with the parameter
imputer_num.fit_transform(missing_na_num)
df_fix[missing_na_num.columns] = imputer_num.fit_transform(df_fix[missing_na_num.columns])

# What will be the number impute in the na value 
imputer_num.statistics_
```

```python
# impute the cat column with the parameter
imputer_cat.fit(missing_na_cat)
df_fix[missing_na_cat.columns] = imputer_cat.transform(df_fix[missing_na_cat.columns])

# What will be the word impute in the na value 
imputer_cat.statistics_
```

```python
# See the final Result: No more Missing Value!
df_fix.isna().sum()
```

```python
# Checking outliers in Numerical columns in df_fix
numerical_cols_fix = df_fix.select_dtypes(['int','float'])
numerical_cols_fix
```

```python
# Finding 1st and 3rd quantile and IQR (helps find outliers)
twenty_fifth = numerical_cols_fix.quantile(0.25)
seventy_fifth = numerical_cols_fix.quantile(0.75)

iqr = seventy_fifth - twenty_fifth
iqr
```

```python
outliers = df_fix[(df_fix[numerical_cols_fix.columns] > right_off)|(df_fix[numerical_cols_fix.columns] < left_off)]
outliers

# Summarize the variables with outliers
```

```python
scalar.fit(df_machine_learning.loc[:,not_dummy_num_col])
```

```python
# All the value has within range from 0 to 1 expect Salesprice
df_machine_learning[not_dummy_num_col] = scalar.transform(df_machine_learning.loc[:,not_dummy_num_col])
df_machine_learning
```

```python
# Collinearity
```

```python
# Take a quick look at the heatmap, we find there are some variables has multiCollinearity with other features
non_dummy_corr = df_machine_learning[not_dummy_num_col].corr()
sns.heatmap(non_dummy_corr, annot= False, cmap='coolwarm')
```

```python
# We use VIF to determine which feature we want to keep
def calculate_vif(df):
    vif_data = pd.DataFrame()
    vif_data["feature"] = df.columns
    vif_data["VIF"] = [1000 if i == 'inf' else variance_inflation_factor(df.values, i) for i in range(len(df.columns))]
    return vif_data

# Assuming df_machine_learning is your DataFrame and not_dummy_num_col is defined
vif_df = calculate_vif(df_machine_learning[not_dummy_num_col])
vif_df.sort_values('VIF', ascending=False, inplace=True)
print(vif_df)
```

```python
#Comments on the features and their corresponding VIF Numbers 
    #Features with high VIF numbers are also very concerning. Their high VIF values mean that they are highly predictable based on other features in the dataset 
    #Features with VIF just above the common threshhold of 5 (SalePrice,FullBath,YearBuilt etc)although less extreme still indicate multicollinearity that could be problematic 
    #Features with VIF values close to 1 are not a concern regarding multicolinearity 
```

```python
# Filter the DataFrame to show only rows with VIF greater than 5
high_vif_features = vif_df[vif_df['VIF'] > 5]

# Creating a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='VIF', y='feature', data=high_vif_features.sort_values('VIF', ascending=False))
plt.title('VIF Values for Features with VIF > 5')
plt.xlabel('Variance Inflation Factor (VIF)')
plt.ylabel('Features')
plt.show()
```

```python
high_vif_features['feature'].to_list()
```

```python
# The first 4 columns has infinity value, we should remove these value to avoid multi-Collinarity
#The bar plot displays that several features have extremely high VIF values 
#This suggests that certain predictor variables are almost perfect linear functions of other predictor variables 
    #(perfectly m ulticollinear)
#This is a strong sign that these variables need to be modified, combined or removed
# We use feature Engineering to remove these columns
df_machine_learning.drop(['LowQualFinSF',
 '1stFlrSF',
 'GrLivArea',
 '2ndFlrSF',
 'TotalBsmtSF',
 'YearBuilt',
 'TotRmsAbvGrd',
 'OverallQual',
 'GarageYrBlt',
 'GarageCars',
 'GarageArea',
 'FullBath',
 'BedroomAbvGr',
 'BsmtUnfSF',
 'KitchenAbvGr',
 'BsmtFinSF1',
 'YearRemodAdd'], axis =1, inplace= True)
```
